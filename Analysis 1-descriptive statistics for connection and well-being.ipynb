```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sheet_url = "https://docs.google.com/spreadsheets/d/1ahcgoAM9zhprLZLWyEFdonzngC5suRUg/export?format=csv"
data = pd.read_csv(sheet_url)

print("Data Preview:")
print(data.head())

for column in data.select_dtypes(include=['float64', 'int64']).columns:
    print(f"\nStatistics for {column}:")
    print(f"Mean: {data[column].mean()}")
    print(f"Median: {data[column].median()}")
    print(f"Mode: {data[column].mode()[0]}")
    print(f"Variance: {data[column].var()}")
    print(f"Standard Deviation: {data[column].std()}")

    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], kde=True)
    plt.title(f'Frequency Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.boxplot(y=data[column])
    plt.title(f'Box Plot of {column}')
    plt.ylabel(column)
    plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

sheet_url = "https://docs.google.com/spreadsheets/d/1HW6CutGBitRQkE80xtC-jMNQdt9sQqE54ZW66x5htB8/export?format=csv"
data = pd.read_csv(sheet_url)

print("Data Preview:")
print(data.head())

question_column = 'CONNECTION_activities_phone_p3m' 

option_freq = data[question_column].value_counts()
print("\nOption Frequency Distribution:\n", option_freq)

option_percentage = data[question_column].value_counts(normalize=True) * 100
print("\nOption Percentage Distribution:\n", option_percentage)

option_mode = data[question_column].mode()[0]
print("\nMost Frequent Option (Mode):", option_mode)

plt.figure(figsize=(10, 6))
option_freq.plot(kind='bar')
plt.title("Option Frequency Distribution")
plt.xlabel("Options")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 8))
option_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title("Option Percentage Distribution")
plt.ylabel("")  
plt.show()
```
